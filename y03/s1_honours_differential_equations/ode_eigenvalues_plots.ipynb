Proper node (two independent eigenvectors)

```python
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt

def sys1(x, t):
    x1, x2 = x
    dydt = []  # ODE system
    return dydt

# Initial conditions
x0 = [1.0, 0.0]  # initial position and velocity
x1 = [-5, 0]

# Time grid
t = np.linspace(0, 10, 1000)

# Solve the ODE
sol1 = odeint(sys1, x0, t)
sol2 = odeint(sys1, x1, t)

# Plot the phase diagram
plt.figure(figsize=(6, 6))
plt.plot(sol1[:, 0], sol1[:, 1])
plt.plot(sol2[:, 0], sol2[:, 1])
plt.title('')
plt.xlabel(r'$x_1$')
plt.ylabel(r"$x_2$")
plt.grid(True)
plt.show()
```
