Proper node (two independent eigenvectors)

```python
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt

def sys1(x, t):
    x1, x2 = x
    dxdt = [x1-x2,x1+3*x2]
    return dxdt

# Initial conditions
x0 = [1, 1]  # initial position and velocity
x1 = [1, -1]
x2 = [-1, -1]

# Time grid
t = np.linspace(0, 10, 10000)

# Solve the ODE
sol1 = odeint(sys1, x0, t)
sol2 = odeint(sys1, x1, t)
sol3 = odeint(sys1, x2, t)

# Plot the phase diagram
plt.figure(figsize=(6, 6))
plt.plot(sol1[:, 0], sol1[:, 1])
plt.plot(sol2[:, 0], sol2[:, 1])
plt.plot(sol3[:, 0], sol3[:, 1])
plt.title('')
plt.xlabel(r'$x_1$')
plt.ylabel(r"$x_2$")
plt.grid(True)
#plt.xlim([-1,1])
#plt.ylim([-1,1])
plt.show()
```
